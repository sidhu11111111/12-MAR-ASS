#QNO.1 ANS

```python
import scipy.stats as stats
sample_mean = 50
sample_std=5
sample_size =100
confidence_interval = 0.95
alpha = 1 - confidence_interval

standard_error = sample_std/(sample_size**0.5)

margin_of_error = stats.t.ppf(1-alpha/2,sample_size)*standard_error

confidence_interval = (sample_mean- margin_of_error, sample_mean +margin_of_error)

print("95% Confidence Interval:", confidence_interval)
```

#QNO.2 ANS

```python
import scipy.stats as stats
import numpy as np

observed_freq = np.array([30, 25, 20, 10, 10, 15]) 
expected_freq = np.array([0.2, 0.2, 0.2, 0.1, 0.1, 0.2]) * np.sum(observed_freq)  

chi2_stat, p_value = stats.chisquare(f_obs=observed_freq, f_exp=expected_freq)

significance_level = 0.05

if p_value < significance_level:
    print("Reject null hypothesis: The distribution of colors does not match the expected distribution.")
else:
    print("Fail to reject null hypothesis: The distribution of colors matches the expected distribution.")
    
print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)

```

#QNO.3 ANS

```python
import scipy.stats as stats

contingency_table = [[20, 15],
                    [10, 25],
                    [15, 20]]
chi2_stat, p_value, _, _ = stats.chi2_contingency(contingency_table)

significance_level = 0.05

if p_value < significance_level:
    print("Reject null hypothesis: There is a significant association between the variables.")
else:
    print("Fail to reject null hypothesis: There is no significant association between the variables.")

print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)
```

Interpreting the results: The chi-square test for a contingency table examines whether there is an association between two categorical variables. The null hypothesis assumes that there is no association, while the alternative hypothesis suggests an association exists.

#QNO.4 ANS

```python
import statsmodels.stats.proportion as proportion

sample_size = 500
smokers = 60

conf_interval = proportion.proportion_confint(smokers, sample_size, alpha = 0.05, method='normal')
print("95% Confidence Interval:",conf_interval)
```

Interpreting the results: The calculated 95% confidence interval represents the range of values within which we can be 95% confident that the true proportion of individuals in the population who smoke lies. The confidence interval is expressed as (lower bound, upper bound). For example, if the calculated interval is (0.095, 0.165), we can say that we are 95% confident that the true proportion of smokers in the population falls within this range.

#QNO.5 ANS

```python
import scipy.stats as stats

sample_mean = 75
sample_std = 12
sample_size = 100  

confidence_level = 0.90
alpha = 1 - confidence_level

standard_error = sample_std / (sample_size ** 0.5)

margin_of_error = stats.t.ppf(1 - alpha / 2, sample_size - 1) * standard_error

confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("90% Confidence Interval:", confidence_interval)
```

Interpreting the results: The calculated 90% confidence interval represents the range of values within which we can be 90% confident that the true population mean lies. The confidence interval is expressed as (lower bound, upper bound). For example, if the calculated interval is (71.68, 78.32), we can say that we are 90% confident that the true population mean falls within this range.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

df= 10
x = np.linspace(0,30,500)

plt.plot(x, stats.chi2.pdf(x,df))

plt.fill_between(x, stats.chi2.pdf(x, df), where =(x >+ 15),color='gray', alpha = 0.5)

plt.xlabel('Chi-square Value')
plt.ylabel('Probability Density')

plt.show()
```

#QNO.8 ANS

```python
import scipy.stats as stats

observed_freq = [45, 55] 
expected_freq = [0.5, 0.5] * 100  

chi2_stat, p_value = stats.chisquare(f_obs=observed_freq, f_exp=expected_freq[:len(observed_freq)])

significance_level = 0.05

if p_value < significance_level:
    print("Reject null hypothesis: The observed frequencies do not match the expected frequencies of a fair coin.")
else:
    print("Fail to reject null hypothesis: The observed frequencies match the expected frequencies of a fair coin.")

print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)

```
